```python
import numpy as np
import matplotlib.pyplot as plt

ti = -np.pi #tiempo inicial
tf = np.pi #tiempo final
wo = 2*np.pi/(tf-ti) #frecuencia angular
Fo = wo/2*np.pi #frecuencia fundamental
print('wo=',wo)
print('Fo=',Fo)
Fs = 10*Fo #frecuencia muestreo -> debe respetar Nyquist Fs >= 2 Fo
tv = np.arange(ti,tf,1/Fs) #vector de tiempo


x = abs(6*np.sin((3*tv)+(np.pi/4)))**2 #señal de entrada
#x = (18+18*np.sin(6*tv)) #señal de entrada
plt.plot(tv,x) #gráfica de la señal
plt.xlabel('tiempo [s]')
plt.grid()
plt.ylabel('')
plt.show()
```

Este código de Python se utiliza para generar y graficar una señal.

1. **Importación de las bibliotecas necesarias**: La biblioteca `numpy` se utiliza para operaciones numéricas y la biblioteca `matplotlib.pyplot` se utiliza para graficar los datos.

2. **Configuración de los parámetros**:
    - `ti` y `tf` son los tiempos inicial y final, respectivamente, establecidos en -π y π.
    - `wo` es la frecuencia angular, calculada como 2π dividido por la diferencia entre el tiempo final e inicial.
    - `Fo` es la frecuencia fundamental, calculada como `wo` dividido por 2π.
    - `Fs` es la frecuencia de muestreo, que debe respetar el teorema de Nyquist (Fs >= 2 Fo). Se establece en 10 veces la frecuencia fundamental.
    - `tv` es un vector de tiempo, creado usando la función `arange` de numpy con un tamaño de paso de 1/Fs.

3. **Generación de la señal**: La señal `x` se genera usando la ecuación |6sin(3t + π/4)|².

4. **Graficación de la señal**: Luego se grafica la señal en función del tiempo usando la función `plot` de matplotlib. El eje x se etiqueta como 'tiempo [s]' y se añade una cuadrícula al gráfico para una mejor visualización.

5. **Visualización del gráfico**: Finalmente, se muestra el gráfico usando la función `show` de matplotlib.

Este código básicamente crea una señal que varía en el tiempo y la grafica durante un intervalo de tiempo especificado. La señal en sí misma es una onda sinusoidal al cuadrado con un desplazamiento de fase, y se toma su valor absoluto para asegurar que siempre sea positiva.

```python
#se crean las bases de la serie exponencial compleja de Fourier
T = tf - ti #intervalo de interés
N = 20 #número armonicos a simular
wo = 2*np.pi/T # frecuencia fundamental
# definir bases
phin = np.zeros((len(tv),2*N+1),dtype=np.complex_)#crear matriz para guardar bases
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)#base de fourier en el intervalo de interés
```

Este código de Python se utiliza para crear las bases de la serie exponencial compleja de Fourier.

1. **Definición de parámetros**:
    - `T` es el intervalo de interés, que se calcula como la diferencia entre el tiempo final `tf` y el tiempo inicial `ti`.
    - `N` es el número de armónicos a simular, que se establece en 20.
    - `wo` es la frecuencia fundamental, que se calcula como 2π dividido por `T`.

2. **Creación de las bases**:
    - Se crea una matriz `phin` para guardar las bases. Esta matriz tiene una longitud igual a la del vector de tiempo `tv` y una anchura igual a dos veces el número de armónicos más uno. Se utiliza el tipo de datos complejo para permitir valores imaginarios.
    - Luego, para cada valor de `n` en el rango desde `-N` hasta `N`, se calcula la base correspondiente utilizando la ecuación e^(inwo*tv) y se guarda en la matriz `phin`. Esta ecuación representa una base de Fourier en el intervalo de interés.

En resumen, este código genera las bases para una serie exponencial compleja de Fourier en un intervalo de tiempo especificado.

```python
# Importación de bibliotecas necesarias
from ipywidgets import interact, IntSlider
import matplotlib.pyplot as plt

# Definición de la función que grafica las bases de Fourier
def pltbase(n=1):
    # Grafica la parte real de la base
    plt.plot(tv, np.real(phin[:, int(n + N)]), label="Re{(t)}")

    # Etiquetas de los ejes y título
    plt.xlabel("t[s]", fontsize=14)
    plt.ylabel("(t)", fontsize=14)

    # Grafica la parte imaginaria de la base
    plt.plot(tv, np.imag(phin[:, int(n + N)]), label="Im{(t)}")

    # Agrega una cuadrícula al gráfico
    plt.grid()

    # Agrega una leyenda para indicar las partes real e imaginaria
    plt.legend()

    # Muestra el gráfico
    plt.show()
    return
# Creación de la interfaz interactiva utilizando interact
interact(pltbase, n=(-N, N, 1))
```

Este código de Python se utiliza para graficar las bases de Fourier de manera interactiva.

1. **Importación de las bibliotecas necesarias**: Se importan las bibliotecas `interact` e `IntSlider` de `ipywidgets` para crear una interfaz interactiva, y `matplotlib.pyplot` para graficar los datos.

2. **Definición de la función que grafica las bases de Fourier (`pltbase`)**: Esta función toma un argumento `n`, que es el índice de la base de Fourier a graficar, y realiza lo siguiente:
    - Grafica la parte real de la base utilizando la función `plot` de matplotlib. La base se obtiene de la matriz `phin` utilizando el índice `n + N`.
    - Establece las etiquetas de los ejes x e y.
    - Grafica la parte imaginaria de la base.
    - Agrega una cuadrícula al gráfico para facilitar la visualización.
    - Agrega una leyenda para indicar cuál línea corresponde a la parte real y cuál a la parte imaginaria.
    - Muestra el gráfico utilizando la función `show` de matplotlib.

3. **Creación de la interfaz interactiva**: Se utiliza la función `interact` para crear una interfaz interactiva que permite al usuario seleccionar el valor de `n` utilizando un control deslizante. El rango del control deslizante va desde `-N` hasta `N`, con un paso de 1.

En resumen, este código permite al usuario seleccionar interactivamente una base de Fourier y graficar tanto su parte real como su parte imaginaria.

```python
# Inicialización de un vector de coeficientes complejos
cn = np.zeros(2*N+1, dtype=np.complex_)

# Creación de un vector de armónicos que abarca desde -N hasta N
nv = np.linspace(-N, N, 2*N+1)

# Asignación de coeficientes específicos
cn[N] = 18       # Componente DC (C_0)
cn[N-6] = 1j*9   # Componente armónico +9j
cn[N+6] = 1j*-9     # Componente armónico -9j
```

Este código de Python se utiliza para inicializar un vector de coeficientes complejos y asignar valores específicos a ciertos coeficientes.
1. **Inicialización de un vector de coeficientes complejos (`cn`)**: Se crea un vector de ceros con una longitud igual a dos veces el número de armónicos más uno. Se utiliza el tipo de datos complejo para permitir valores imaginarios.

2. **Creación de un vector de armónicos (`nv`)**: Se utiliza la función `linspace` de numpy para crear un vector que va desde `-N` hasta `N`, con una longitud total de `2*N+1`.

3. **Asignación de coeficientes específicos**: Se asignan valores específicos a ciertos coeficientes en el vector `cn`.
    - El coeficiente en el índice `N` (que corresponde a la componente DC o C_0) se establece en 18.
    - El coeficiente en el índice `N-6` (que corresponde al armónico -6) se establece en `1j*9`, donde `1j` representa la unidad imaginaria.
    - El coeficiente en el índice `N+6` (que corresponde al armónico +6) se establece en `1j*-9`.

En resumen, este código inicializa un vector para almacenar los coeficientes de una serie de Fourier y asigna valores específicos a ciertos coeficientes.

```python
# Creación de una figura para los gráficos
fig = plt.figure(figsize=(8, 8))

# Gráfico de la parte real de los coeficientes Cn
plt.subplot(2, 2, 1)
plt.stem(nv, np.real(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Re{Cn}', fontsize=14)
plt.axis('tight')
plt.grid()

# Gráfico de la parte imaginaria de los coeficientes Cn
plt.subplot(2, 2, 2)
plt.stem(nv, np.imag(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Im{Cn}', fontsize=14)
plt.axis('tight')
plt.grid()

# Gráfico de la magnitud del espectro (abs(Cn))
plt.subplot(2, 2, 3)
plt.stem(nv, abs(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Cn', fontsize=14)
plt.axis('tight')
plt.grid()

# Gráfico de la fase del espectro (angle(Cn))
plt.subplot(2, 2, 4)
plt.stem(nv, np.angle(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Cn', fontsize=14)
plt.axis('tight')
plt.grid()

# Ajusta el diseño de los gráficos para una mejor presentación
fig.tight_layout()
```

Este código de Python se utiliza para crear una figura y graficar los coeficientes de Fourier.

1. **Creación de una figura para los gráficos**: Se utiliza la función `figure` de matplotlib para crear una nueva figura con un tamaño específico.

2. **Gráfico de la parte real de los coeficientes Cn**: Se utiliza la función `subplot` para crear un subgráfico en la primera posición de una cuadrícula 2x2. Luego, se utiliza la función `stem` para crear un gráfico de tallo y hojas de la parte real de los coeficientes. Se añaden etiquetas a los ejes, se ajusta el eje para que se ajuste a los datos y se añade una cuadrícula.

3. **Gráfico de la parte imaginaria de los coeficientes Cn**: De manera similar, se crea un subgráfico en la segunda posición y se grafica la parte imaginaria de los coeficientes.

4. **Gráfico de la magnitud del espectro (abs(Cn))**: En el tercer subgráfico, se grafica la magnitud (valor absoluto) de los coeficientes.

5. **Gráfico de la fase del espectro (angle(Cn))**: En el cuarto y último subgráfico, se grafica el ángulo (fase) de los coeficientes.

6. **Ajuste del diseño de los gráficos**: Finalmente, se utiliza la función `tight_layout` para ajustar automáticamente el diseño de los subgráficos dentro de la figura para una mejor presentación.

En resumen, este código crea una figura con cuatro subgráficos que muestran diferentes aspectos de los coeficientes de Fourier: su parte real, su parte imaginaria, su magnitud y su fase.

```python
# Potencia de la señal original x(t)
Px = 486

# Definición de la función que grafica la reconstrucción de la señal
def pltest(Na=1):
    # Rango de índices de coeficientes a considerar
    ind = range(N - Na, N + Na + 1)

    # Cálculo del error de reconstrucción
    er = 1 - np.sum(abs(cn[ind])**2) / Px

    # Señal reconstruida (xe) utilizando las bases y coeficientes seleccionados
    xe = phin[:, ind].dot(cn[ind])  # La función dot calcula el producto entre matrices/vectores

    # Graficar la señal estimada (en azul) y la señal original (en rojo)
    plt.plot(tv, xe, color='b', label='x(t)')  # Señal estimada o filtrada
    plt.plot(tv, x, color='r', label='x(t)')  # Señal original

    # Título del gráfico que muestra el error de reconstrucción
    plt.title('E_r=%.2f[%%]' % (100 * er), fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()
    return

# Creación de una interfaz interactiva para ajustar el número de armónicos utilizados (Na)
interact(pltest, Na=(1, N, 1))
```

Este código de Python se utiliza para reconstruir una señal a partir de sus coeficientes de Fourier y graficar tanto la señal reconstruida como la original.

1. **Definición de la potencia de la señal original (`Px`)**: Se establece en 486.

2. **Definición de la función que grafica la reconstrucción de la señal (`pltest`)**: Esta función toma un argumento `Na`, que es el número de armónicos a considerar para la reconstrucción, y realiza lo siguiente:
    - Calcula el rango de índices de los coeficientes a considerar.
    - Calcula el error de reconstrucción como 1 menos la suma del cuadrado del valor absoluto de los coeficientes seleccionados dividido por `Px`.
    - Reconstruye la señal (`xe`) utilizando las bases y coeficientes seleccionados. Esto se hace utilizando la función `dot` para calcular el producto entre las bases y los coeficientes.
    - Grafica tanto la señal reconstruida (en azul) como la señal original (en rojo). También muestra el error de reconstrucción en el título del gráfico.

3. **Creación de una interfaz interactiva**: Se utiliza la función `interact` para crear una interfaz interactiva que permite al usuario seleccionar el valor de `Na` utilizando un control deslizante. El rango del control deslizante va desde 1 hasta `N`, con un paso de 1.

En resumen, este código permite al usuario seleccionar cuántos armónicos se deben considerar para reconstruir una señal a partir de sus coeficientes de Fourier, y luego grafica tanto la señal reconstruida como la original.

Este código de Python se utiliza para calcular y graficar el espectro de Fourier de una señal de audio.

La función `rfft` se utiliza específicamente para señales reales y devuelve la mitad del espectro, ya que la otra mitad es simétrica para señales reales.

2. **`vf = np.fft.rfftfreq(m.shape[0],1/fs)`**: Esta línea crea un vector de frecuencias `vf` que corresponde a las frecuencias de los componentes en el espectro de Fourier. Se utiliza la función `rfftfreq` del módulo `np.fft`, que toma como argumentos el número de muestras en la señal y el período de muestreo (que es el recíproco de la frecuencia de muestreo `fs`).

3. **`plt.plot(vf,abs(Mw))`**: Esta línea grafica la magnitud del espectro de Fourier en función de la frecuencia. Se utiliza la función `abs` para calcular la magnitud del espectro, que es un número complejo.

En resumen, este código calcula el espectro de Fourier de una señal de audio y lo grafica en función de la frecuencia.

```python
Fc = 15000 #frecuencia portadora
Im = 1 # Im = peak(m(t))/Ac ->  Ac = peak(m(t))/Im
Ac = max(abs(m))/Im #amplitud de la portadora según índice de modulación
c = Ac*np.cos(2*np.pi*Fc*tt) #portadora
```

Este código de Python se utiliza para generar una señal portadora para la modulación de amplitud.
1. **`Fc = 15000`**: Esta línea establece la frecuencia de la señal portadora en 15000 Hz.

2. **`Im = 1`**: Esta línea establece el índice de modulación, que es la relación entre la amplitud máxima de la señal modulante y la amplitud de la señal portadora.

3. **`Ac = max(abs(m))/Im`**: Esta línea calcula la amplitud de la señal portadora. Se hace dividiendo la amplitud máxima de la señal modulante (obtenida con `max(abs(m))`) por el índice de modulación.

4. **`c = Ac*np.cos(2*np.pi*Fc*tt)`**: Finalmente, esta línea genera la señal portadora `c`. Es una onda coseno con una frecuencia igual a `Fc` y una amplitud igual a `Ac`.

En resumen, este código genera una señal portadora para ser utilizada en la modulación de amplitud.
