```python
#Exercises 4_IntroNumpy
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

#Datos señal continua
A=120 #amplitud
f=60  #frecuencia
T=1/f #periodo

#Discretizar tiempo
ti=0 #tiempo inicial
tf=T #tiempo final
Fs=50*f #frecuencia de muestreo
Ts=1/Fs #tiempo de muestreo

#Crear vector discretizado
tv=np.arange(ti,tf,Ts)
print(tv)
print(tv.shape)

#Discretizar señal
xv=A*np.cos(2*np.pi*f*tv)
print(xv)
print(xv.shape)

#imprimir señal discretizada
plt.stem(tv,xv)
plt.show()

#Acondicionar la señal (modificar amplitud)
y_max=5
y_min=0
def modificar_amplitud (x,ymin,ymax):
  #Vamos a hallar la pendiente
  xmin=min(x) #Esta función me saca el minimo de la señal discretizada
  xmax=max(x) #Esta función me saca el maximo de la señal discretizada
  m=(ymax-ymin)/(xmax-xmin) #Aquí modificamos la pendiente para escalar la señal y que el dispositivo que trabaja de 0 a 5v la reciba de 0 a 5v
  c=ymax-m*xmax #Hallamos la c, porque la ecuación y=mx+c lo requiere y para que se mantenga la forma de la señal original
  yv=m*x+c #aqui calculamos la señal discretizada de 0 a 5v
  return yv

#vamos a pasarle los datos a la función modificar_amplitud para que nos calcule la señal acondicionada (yv)
yv= modificar_amplitud(xv,y_min,y_max)
#imprimir señal acondicionada
plt.plot(tv,yv,"m")
plt.stem(tv,yv)
plt.show()

#Vamos a cuantizar la señal

#primero, creamos los cajones según el numero de bits del dispositivo
nbits=3 #numero de bits
rmin= 0 #estado máximo
rmax= 5 #estado máximo
ve=np.linspace(rmin,rmax,2**nbits) #Se creo un vector (cajones) según el numero de bits del dispositivo
print(ve)
print(ve.shape)

#Ahora, vamos a hallar las distancias entre la señal discretizada (yv) y los cajones (ve)
d= cdist(yv.reshape(-1,1),ve.reshape(-1,1)) #utilizamos el reshape para modificar la dimension del vector al de una matriz y asi calcular las distancias
print(d)
d.shape

#Ahora, vamos a calcular la distancia más pequeña del cualquier punto respecto a un cajón para graficarla
ind=np.argmin(d,axis=1) # el axis=1 significa que va de 1 en 1 en la matriz de distancias encontrando la distancia más cerca
valores_cuantizados=ve[ind] #Aquí, ponemos el valor de yv en el cajon más cercano
print(yv)
print(valores_cuantizados)

#Ahora, vamos a imprimir la señal cuantizada (digital)
plt.plot(tv,valores_cuantizados,'r')
plt.xlabel("tiempo [seg]")
plt.ylabel("señal cuantizada [V]")
plt.show()
```

```python
#Exercises 5_Functions

#1

from math import sqrt # De la librería math, importo sqrt que es una función que me calcula la raiz cuadrada
def area (s1,s2,s3): #definimos una función de nombre área que recibe s1, s2 y s3 como parametros por pocsición
  s= (s1+s2+s3)/2 #calculamos el subperimetro
  area=sqrt(s*(s-s1)*(s-s2)*(s-s3)) #calculo del area del triángulo
  area_redondeada= round(area,1) #creo una variable llamada area redondeada y utilizamos la función round, le damos como parametros el numero a redondear y el numero de decimales al cual queremos redondear
  return area_redondeada #me retorna el área

s1=float(input("Ingrese el lado 1 del triangulo: ")) #pregunto por el lado 1 del triangulo y lo almaceno
s2=float(input("Ingrese el lado 2 del triangulo: ")) #pregunto por el lado 2 del triangulo y lo almaceno
s3=float(input("Ingrese el lado 3 del triangulo: ")) #pregunto por el lado 3 del triangulo y lo almaceno

area = area(s1,s2,s3) #defino una variable llamada area, llamo la función area y le paso los parametros por posición
print("El area del triangulo es: ",area) #imprimo en pantalla el area calculada

#2

def BMI (peso,altura): #definimos una función de nombre BMI que recibe peso y altura como parametros por posición
  peso_kg=peso*0.45 #convertimos el peso de libras a kilogramos
  altura_mts=altura*0.025 #convertimos la altura de pulgadas a metros
  BMI=peso_kg/altura_mts**2
  BMI_redondeado= round(BMI,2) #creo una variable llamada area redondeada y utilizamos la función round, le damos como parametros el numero a redondear y el numero de decimales al cual queremos redondear
  return BMI_redondeado #me retorna el BMI

peso=float(input("Ingrese el peso en libras : ")) #pregunto por el peso en libras y lo almaceno
altura=float(input("Ingrese la altura en pulgadas: ")) #pregunto por la altura en pulgadas y lo almaceno


BMI = BMI(peso,altura) #defino una variable llamada BMI, llamo la función BMI y le paso los parametros por posición
print("El BMI es: ",BMI) #imprimo en pantalla el BMI calculada
```

```python
#Exercises 8_Herencias

#1

import numpy as np

#Creo clase base llamada FiguraGeometrica
class FiguraGeometrica:
    def __init__(self, base, altura): #Defino método constructor, que es para darle un estado inicial a los objetos de esta clase
      self.base = base #atributo base
      self.altura = altura #atributo altura

#Creo clase base llamada FiguraGeometrica_Conica
class FiguraGeometrica_Conica:
    def __init__(self, radio): #Defino método constructor, que es para darle un estado inicial a los objetos de esta clase
      self.radio = radio #atributo radio

class Triangulo(FiguraGeometrica): #Creo una clase derivada de FiguraGeometrica llamada Triangulo
    def __init__(self, base, altura): #Defino método constructor, que es para darle un estado inicial a los objetos de esta clase
      super().__init__(base,altura) #Heredo el método constructor de la clase base

    def CalcularAreaT(self): #Defino método de clase Triangulo llamado CalcularAreaT y le paso como argumento el objeto
      area= round((self.base*self.altura)/2,3) #Calculo del area
      print('El area del triangulo es: ',area) #imprimo el area

class Cuadrilatero(FiguraGeometrica): #Creo una clase derivada de FiguraGeometrica llamada Cuadrilatero
    def __init__(self, base, altura): #Defino método constructor, que es para darle un estado inicial a los objetos de esta clase
      super().__init__(base, altura) #Heredo el método constructor de la clase base

    def CalcularAreaC(self): #Defino método de clase Cuadrilatero llamado CalcularAreaC y le paso como argumento el objeto
      area = self.base*self.altura #Calculo del area
      print('El area del cuadrilatero es:', area ) #imprimo el area

class Circulo(FiguraGeometrica_Conica): #Creo una clase derivada de FiguraGeometrica_Conica llamada Circulo
    def _init_(self, radio): #Defino método constructor, que es para darle un estado inicial a los objetos de esta clase
      super()._init_(radio) #Heredo el método constructor de la clase base

    def CalcularAreaCi(self): #Defino método de clase Circulo llamado CalcularAreaCi y le paso como argumento el objeto
      area = round(np.pi*(self.radio**2), 3) #Calculo del area
      print('El área del circulo es: ', area) #imprimo el area

#Creación de objetos

triangulo1 = Triangulo(20,5)
triangulo2 = Triangulo(10,5)
triangulo3 = Triangulo(10,15)
cuadrilatero = Cuadrilatero(20,5)
circulo = Circulo(4)

#Llamo parametros

triangulo1.CalcularAreaT()
triangulo2.CalcularAreaT()
triangulo3.CalcularAreaT()
cuadrilatero.CalcularAreaC()
circulo.CalcularAreaCi()

#2

asientos = [[1,1,1],[1,1,1],[1,1,1]]
tarifa = 50000


class Vuelo:
    def _init_(self, nvuelo, origen, destino, cmpasajeros):
      self.nvuelo = nvuelo
      self.origen = origen
      self.destino = destino
      self.cmpasajeros = cmpasajeros

    def ocuparAsientos(self):
      print('Asientos donde 1 es disponible y 0 es ocupado\n',asientos)
      contador = 0
      x = int(input('Ingrese el número de asientos que desea comprar: '))
      while contador < x:
        fila = int(input('Seleccion una fila sea 0, 1 o 2: '))
        columna = int(input('Seleccione una columna de 0, 1 o 2: '))
        if asientos[fila][columna] == 1:
          asientos[fila][columna] = 0
          contador += 1
          print('Su asiento esta en la fila: ', fila,'y columna: ',columna)

        else:
          print('Asiento ocupado')
class VueloEconomico(Vuelo):
    def _init_(self, nvuelo, origen, destino, cmpasajeros, tarifa):
      super()._init_(nvuelo, origen, destino, cmpasajeros)
      self.tarifa = tarifa

    def ocuparAsientos(self):
      print('Asientos donde 1 es disponible y 0 es ocupado\n',asientos)
      x = int(input('Cuantos asientos desea reservar: '))
      contador = 0
      while contador < x:
        fila = int(input('Seleccion una fila sea 0, 1 o 2: '))
        columna = int(input('Seleccione una columna de 0, 1 o 2: '))
        if asientos[fila][columna] == 1:
          asientos[fila][columna] = 0
          print('Su asiento esta en la fila: ', fila,'y columna: ',columna)
          contador += 1
        else:
          print('Asiento ocupado')
      precio = self.tarifa*x
      print('La tarifa total es: ', precio)


vuelo = Vuelo(62,'pereira', 'manizales', 7)
vuelo.ocuparAsientos()
vueloe= VueloEconomico(62,'pereira', 'manizales', 7, tarifa)
vueloe.ocuparAsientos()
```
