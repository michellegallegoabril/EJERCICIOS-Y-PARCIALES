```python
import numpy as np
import matplotlib.pyplot as plt

ti = -np.pi #tiempo inicial
tf = np.pi #tiempo final
wo = 2*np.pi/(tf-ti) #frecuencia angular
Fo = wo/2*np.pi #frecuencia fundamental
print('wo=',wo)
print('Fo=',Fo)
Fs = 10*Fo #frecuencia muestreo -> debe respetar Nyquist Fs >= 2 Fo
tv = np.arange(ti,tf,1/Fs) #vector de tiempo


x = abs(6*np.sin((3*tv)+(np.pi/4)))**2 #seal de entrada
#x = (18+18*np.sin(6*tv)) #seal de entrada
plt.plot(tv,x) #grfica de la seal
plt.xlabel('tiempo [s]')
plt.grid()
plt.ylabel('')
plt.show()
```

Este cdigo de Python se utiliza para generar y graficar una seal.

1. **Importacin de las bibliotecas necesarias**: La biblioteca `numpy` se utiliza para operaciones numricas y la biblioteca `matplotlib.pyplot` se utiliza para graficar los datos.

2. **Configuracin de los parmetros**:
    - `ti` y `tf` son los tiempos inicial y final, respectivamente, establecidos en - y .
    - `wo` es la frecuencia angular, calculada como 2 dividido por la diferencia entre el tiempo final e inicial.
    - `Fo` es la frecuencia fundamental, calculada como `wo` dividido por 2.
    - `Fs` es la frecuencia de muestreo, que debe respetar el teorema de Nyquist (Fs >= 2 Fo). Se establece en 10 veces la frecuencia fundamental.
    - `tv` es un vector de tiempo, creado usando la funcin `arange` de numpy con un tamao de paso de 1/Fs.

3. **Generacin de la seal**: La seal `x` se genera usando la ecuacin |6sin(3t + /4)|.

4. **Graficacin de la seal**: Luego se grafica la seal en funcin del tiempo usando la funcin `plot` de matplotlib. El eje x se etiqueta como 'tiempo [s]' y se aade una cuadrcula al grfico para una mejor visualizacin.

5. **Visualizacin del grfico**: Finalmente, se muestra el grfico usando la funcin `show` de matplotlib.

Este cdigo bsicamente crea una seal que vara en el tiempo y la grafica durante un intervalo de tiempo especificado. La seal en s misma es una onda sinusoidal al cuadrado con un desplazamiento de fase, y se toma su valor absoluto para asegurar que siempre sea positiva.

```python
#se crean las bases de la serie exponencial compleja de Fourier
T = tf - ti #intervalo de inters
N = 20 #nmero armonicos a simular
wo = 2*np.pi/T # frecuencia fundamental
# definir bases
phin = np.zeros((len(tv),2*N+1),dtype=np.complex_)#crear matriz para guardar bases
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)#base de fourier en el intervalodeinters
```

Este cdigo de Python se utiliza para crear las bases de la serie exponencial compleja de Fourier.

1. **Definicin de parmetros**:
    - `T` es el intervalo de inters, que se calcula como la diferencia entre el tiempo final `tf` y el tiempo inicial `ti`.
    - `N` es el nmero de armnicos a simular, que se establece en 20.
    - `wo` es la frecuencia fundamental, que se calcula como 2 dividido por `T`.

2. **Creacin de las bases**:
    - Se crea una matriz `phin` para guardar las bases. Esta matriz tiene una longitud igual a la del vector de tiempo `tv` y una anchura igual a dos veces el nmero de armnicos ms uno. Se utiliza el tipo de datos complejo para permitir valores imaginarios.
    - Luego, para cada valor de `n` en el rango desde `-N` hasta `N`, se calcula la base correspondiente utilizando la ecuacin e^(inwo*tv) y se guarda en la matriz `phin`. Esta ecuacin representa una base de Fourier en el intervalo de inters.

En resumen, este cdigo genera las bases para una serie exponencial compleja de Fourier en un intervalo de tiempo especificado.

```python
# Importacin de bibliotecas necesarias
from ipywidgets import interact, IntSlider
import matplotlib.pyplot as plt

# Definicin de la funcin que grafica las bases de Fourier
def pltbase(n=1):
    # Grafica la parte real de la base
    plt.plot(tv, np.real(phin[:, int(n + N)]), label="Re{(t)}")

    # Etiquetas de los ejes y ttulo
    plt.xlabel("t[s]", fontsize=14)
    plt.ylabel("(t)", fontsize=14)

    # Grafica la parte imaginaria de la base
    plt.plot(tv, np.imag(phin[:, int(n + N)]), label="Im{(t)}")

    # Agrega una cuadrcula al grfico
    plt.grid()

    # Agrega una leyenda para indicar las partes real e imaginaria
    plt.legend()

    # Muestra el grfico
    plt.show()
    return
# Creacin de la interfaz interactiva utilizando interact
interact(pltbase, n=(-N, N, 1))
```

    - Grafica la parte real de la base utilizando la funcin `plot` de matplotlib. La base se obtiene de la matriz `phin` utilizando el ndice `n + N`.
    - Establece las etiquetas de los ejes x e y.
    - Grafica la parte imaginaria de la base.
    - Agrega una cuadrcula al grfico para facilitar la visualizacin.
    - Agrega una leyenda para indicar cul lnea corresponde a la parte real y cul a la parte imaginaria.
    - Muestra el grfico utilizando la funcin `show` de matplotlib.

3. **Creacin de la interfaz interactiva**: Se utiliza la funcin `interact` para crear una interfaz interactiva que permite al usuario seleccionar el valor de `n` utilizando un control deslizante. El rango del control deslizante va desde `-N` hasta `N`, con un paso de 1.

En resumen, este cdigo permite al usuario seleccionar interactivamente una base de Fourier y graficar tanto su parte real como su parte imaginaria.

```python
# Inicializacin de un vector de coeficientes complejos
cn = np.zeros(2*N+1, dtype=np.complex_)

# Creacin de un vector de armnicos que abarca desde -N hasta N
nv = np.linspace(-N, N, 2*N+1)

# Asignacin de coeficientes especficos
cn[N] = 18       # Componente DC (C_0)
cn[N-6] = 1j*9   # Componente armnico +9j
cn[N+6] = 1j*-9     # Componentearmnico-9j
```

Este cdigo de Python se utiliza para inicializar un vector de coeficientes complejos y asignar valores especficos a ciertos coeficientes.
1. **Inicializacin de un vector de coeficientes complejos (`cn`)**: Se crea un vector de ceros con una longitud igual a dos veces el nmero de armnicos ms uno. Se utiliza el tipo de datos complejo para permitir valores imaginarios.

2. **Creacin de un vector de armnicos (`nv`)**: Se utiliza la funcin `linspace` de numpy para crear un vector que va desde `-N` hasta `N`, con una longitud total de `2*N+1`.

3. **Asignacin de coeficientes especficos**: Se asignan valores especficos a ciertos coeficientes en el vector `cn`.
    - El coeficiente en el ndice `N` (que corresponde a la componente DC o C_0) se establece en 18.
    - El coeficiente en el ndice `N-6` (que corresponde al armnico -6) se establece en `1j*9`, donde `1j` representa la unidad imaginaria.
    - El coeficiente en el ndice `N+6` (que corresponde al armnico +6) se establece en `1j*-9`.

En resumen, este cdigo inicializa un vector para almacenar los coeficientes de una serie de Fourier y asigna valores especficos a ciertos coeficientes.

```python
# Creacin de una figura para los grficos
fig = plt.figure(figsize=(8, 8))

# Grfico de la parte real de los coeficientes Cn
plt.subplot(2, 2, 1)
plt.stem(nv, np.real(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Re{Cn}', fontsize=14)
plt.axis('tight')
plt.grid()

# Grfico de la parte imaginaria de los coeficientes Cn
plt.subplot(2, 2, 2)
plt.stem(nv, np.imag(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Im{Cn}', fontsize=14)
plt.axis('tight')
plt.grid()

# Grfico de la magnitud del espectro (abs(Cn))
plt.subplot(2, 2, 3)
plt.stem(nv, abs(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Cn', fontsize=14)
plt.axis('tight')
plt.grid()

# Grfico de la fase del espectro (angle(Cn))
plt.subplot(2, 2, 4)
plt.stem(nv, np.angle(cn), 'r')
plt.xlabel(r'nw [rad/s]', fontsize=14)
plt.ylabel(r'Cn', fontsize=14)
plt.axis('tight')
plt.grid()

# Ajusta el diseo de los grficos para una mejor presentacin
fig.tight_layout()
```

Este cdigo de Python se utiliza para crear una figura y graficar los coeficientes de Fourier.

1. **Creacin de una figura para los grficos**: Se utiliza la funcin `figure` de matplotlib para crear una nueva figura con un tamao especfico.

2. **Grfico de la parte real de los coeficientes Cn**: Se utiliza la funcin `subplot` para crear un subgrfico en la primera posicin de una cuadrcula 2x2. Luego, se utiliza la funcin `stem` para crear un grfico de tallo y hojas de la parte real de los coeficientes. Se aaden etiquetas a los ejes, se ajusta el eje para que se ajuste a los datos y se aade una cuadrcula.

3. **Grfico de la parte imaginaria de los coeficientes Cn**: De manera similar, se crea un subgrfico en la segunda posicin y se grafica la parte imaginaria de los coeficientes.

4. **Grfico de la magnitud del espectro (abs(Cn))**: En el tercer subgrfico, se grafica la magnitud (valor absoluto) de los coeficientes.

5. **Grfico de la fase del espectro (angle(Cn))**: En el cuarto y ltimo subgrfico, se grafica el ngulo (fase) de los coeficientes.

6. **Ajuste del diseo de los grficos**: Finalmente, se utiliza la funcin `tight_layout` para ajustar automticamente el diseo de los subgrficos dentro de la figura para una mejor presentacin.

En resumen, este cdigo crea una figura con cuatro subgrficos que muestran diferentes aspectos de los coeficientes de Fourier: su parte real, su parte imaginaria, su magnitud y su fase.

```python
# Potencia de la seal original x(t)
Px = 486

# Definicin de la funcin que grafica la reconstruccin de la seal
def pltest(Na=1):
    # Rango de ndices de coeficientes a considerar
    ind = range(N - Na, N + Na + 1)

    # Clculo del error de reconstruccin
    er = 1 - np.sum(abs(cn[ind])**2) / Px

    # Seal reconstruida (xe) utilizando las bases y coeficientes seleccionados
    xe = phin[:, ind].dot(cn[ind])  # La funcin dot calcula el producto entre matrices/vectores

    # Graficar la seal estimada (en azul) y la seal original (en rojo)
    plt.plot(tv, xe, color='b', label='x(t)')  # Seal estimada o filtrada
    plt.plot(tv, x, color='r', label='x(t)')  # Seal original

    # Ttulo del grfico que muestra el error de reconstruccin
    plt.title('E_r=%.2f[%%]' % (100 * er), fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()
    return

# Creacin de una interfaz interactiva para ajustar el nmero de armnicos utilizados (Na)
interact(pltest, Na=(1, N, 1))
```

Este cdigo de Python se utiliza para reconstruir una seal a partir de sus coeficientes de Fourier y graficar tanto la seal reconstruida como la original.

1. **Definicin de la potencia de la seal original (`Px`)**: Se establece en 486.

2. **Definicin de la funcin que grafica la reconstruccin de la seal (`pltest`)**: Esta funcin toma un argumento `Na`, que es el nmero de armnicos a considerar para la reconstruccin, y realiza lo siguiente:
    - Calcula el rango de ndices de los coeficientes a considerar.
    - Calcula el error de reconstruccin como 1 menos la suma del cuadrado del valor absoluto de los coeficientes seleccionados dividido por `Px`.
    - Reconstruye la seal (`xe`) utilizando las bases y coeficientes seleccionados. Esto se hace utilizando la funcin `dot` para calcular el producto entre las bases y los coeficientes.
    - Grafica tanto la seal reconstruida (en azul) como la seal original (en rojo). Tambin muestra el error de reconstruccin en el ttulo del grfico.

3. **Creacin de una interfaz interactiva**: Se utiliza la funcin `interact` para crear una interfaz interactiva que permite al usuario seleccionar el valor de `Na` utilizando un control deslizante. El rango del control deslizante va desde 1 hasta `N`, con un paso de 1.

En resumen, este cdigo permite al usuario seleccionar cuntos armnicos se deben considerar para reconstruir una seal a partir de sus coeficientes de Fourier, y luego grafica tanto la seal reconstruida como la original.

La funcin `rfft` se utiliza especficamente para seales reales y devuelve la mitad del espectro, ya que la otra mitad es simtrica para seales reales.

2. **`vf = np.fft.rfftfreq(m.shape[0],1/fs)`**: Esta lnea crea un vector de frecuencias `vf` que corresponde a las frecuencias de los componentes en el espectro de Fourier. Se utiliza la funcin `rfftfreq` del mdulo `np.fft`, que toma como argumentos el nmero de muestras en la seal y el perodo de muestreo (que es el recproco de la frecuencia de muestreo `fs`).

3. **`plt.plot(vf,abs(Mw))`**: Esta lnea grafica la magnitud del espectro de Fourier en funcin de la frecuencia. Se utiliza la funcin `abs` para calcular la magnitud del espectro, que es un nmero complejo.

4. **`plt.title(r'Espectro seal mensaje')`, `plt.xlabel(r'$f[Hz]$')`, y `plt.ylabel(r'$|M[f]|$')`**: Estas lneas aaden un ttulo al grfico y etiquetas a los ejes x e y, respectivamente.

Este cdigo de Python se utiliza para generar una seal portadora para la modulacin de amplitud.
1. **`Fc = 15000`**: Esta lnea establece la frecuencia de la seal portadora en 15000 Hz.

2. **`Im = 1`**: Esta lnea establece el ndice de modulacin, que es la relacin entre la amplitud mxima de la seal modulante y la amplitud de la seal portadora.

3. **`Ac = max(abs(m))/Im`**: Esta lnea calcula la amplitud de la seal portadora. Se hace dividiendo la amplitud mxima de la seal modulante (obtenida con `max(abs(m))`) por el ndice de modulacin.

4. **`c = Ac*np.cos(2*np.pi*Fc*tt)`**: Finalmente, esta lnea genera la seal portadora `c`. Es una onda coseno con una frecuencia igual a `Fc` y una amplitud igual a `Ac`.

En resumen, este cdigo genera una seal portadora para ser utilizada en la modulacin de amplitud.
